```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
```

```python
df = pd.read_csv('./GraphDataSet/RNF.csv',  index_col=None, header=None)
df.columns = ['Missing Rate', 'DataSet', 'Performance']
```

```python
df.head()
```

```python
#df = df.pivot('Missing Rate', 'DataSet', 'Performance')
df = df.groupby(['Missing Rate', 'DataSet'])
df = df['Performance'].aggregate(np.sum).unstack()
```

```python
df.plot(kind = 'bar', title = 'Missing Rate and Performance with RnF', figsize=(8, 5))
plt.ylabel('Performance')
plt.xlabel('Missing Rate %')
plt.show()
```

```python
df.plot(kind = 'line', title = 'Missing Rate and Performance with RnF', figsize=(8, 5))
plt.ylabel('Performance')
plt.xlabel('Missing Rate %')
plt.show()
```

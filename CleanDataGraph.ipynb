```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./GraphDataSet/CleanData.csv',  index_col=None, header=None)
df.columns = ['Classifier', 'DataSet', 'Performance']
```

```python
df.head()
```

```python
#df = df.pivot('Missing Rate', 'DataSet', 'Performance')
df = df.groupby(['Classifier', 'DataSet'])
df = df['Performance'].aggregate(np.sum).unstack()
```

```python
df.plot(kind = 'bar', title = 'Clean Data Performance Result', figsize=(8, 5))
plt.legend(loc='upper right')
plt.ylabel('Performance')
plt.xlabel('Classifier')
plt.show()
```

```python
df.plot(kind = 'line', title = 'Clean Data Performance Result', figsize=(8, 5))
plt.ylabel('Performance')
plt.xlabel('Classifier')
plt.show()
```
